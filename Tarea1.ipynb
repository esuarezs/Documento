1. Una ecuación lineal única con tres entradas (es decir, linear_function_3D()) asocia un valor y con cada punto en un espacio tridimensional (x1, x2, x3). ¿Es posible visualizar esto? ¿Qué valor se encuentra en la posición (0,0,0)?


Si es posible visualizar una ecuación lineal con tres entradas representada como:

image.png


Los coeficientes m,n y p son los coeficientes de las variables x1, x2 y x3 respectivamente y b es la constante o intercepto.
La representación gráfica de la cuación anterior sería en un espacio tridimensional con tres ejes o coordenadas x1, x2 y x3.

Acontinuación una visualización de este tipo de ecuaciones.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def linear_function_3D(x1, x2, x3):
    m, n, p, b = 2, 1, -3, 5
    return m * x1 + n * x2 + p * x3 + 0

# Generar puntos en el espacio tridimensional
x1_values = np.linspace(-10, 10, 100)
x2_values = np.linspace(-10, 10, 100)
x3_values = np.linspace(-10, 10, 100)

# Crear una cuadrícula tridimensional
x1, x2, x3 = np.meshgrid(x1_values, x2_values, x3_values)

# Calcular los valores de y utilizando la ecuación lineal
y = linear_function_3D(x1, x2, x3)

# Configurar la figura tridimensional
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Visualizar el plano
ax.plot_surface(x1[:,:,0], x2[:,:,0], y[:,:,0], alpha=0.5, color='c', edgecolors='k')

# Configuración de los ejes
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

# Mostrar la figura
plt.show()
```

cuando x1=0, x2=0 y x3=0 , la ecuación anterios queda de la forma y=b, es decir el intercepto.

2.  Escribir un código que resuelva los valores de un conjunto de de dos ecuaciones con dos vatiables cada una, de dos formas diferentes, a partir de las ecuaciones individuales y usando matrices.

```python
import numpy as np

# Definir valores de entrada
x1_values = np.array([1, 2, 3])
x2_values = np.array([4, 5, 6])

# Definir los coeficientes de las ecuaciones lineales
m1, n1, b1 = 2, 3, 1
m2, n2, b2 = -1, 4, 2
m3, n3, b3 = 3, 1, 5

# Calcular valores usando ecuaciones individuales
y1_individual = m1 * x1_values + n1 * x2_values + b1
y2_individual = m2 * x1_values + n2 * x2_values + b2
y3_individual = m3 * x1_values + n3 * x2_values + b3

# Calcular valores usando forma matricial
coeff_matrix = np.array([[m1, n1], [m2, n2], [m3, n3]])
x_values_matrix = np.column_stack((x1_values, x2_values))
y_matrix = np.dot(coeff_matrix, x_values_matrix.T)

# Imprimir resultados
print("Resultados usando ecuaciones individuales:")
print("y1:", y1_individual)
print("y2:", y2_individual)
print("y3:", y3_individual)

print("\nResultados usando forma matricial:")
print("y_matrix:", y_matrix)
```
